**Colormaps in Matplotlib**
---------------------------------------

**Matplotlib?** 

Matplotlib는 파이썬에서 자료를 차트(chart)나 플롯(plot)으로 시각화(visulaization)하는 패키지.

Matplotlib는 다음과 같은 정형화된 차트나 플롯 이외에도 저수준 api를 사용한 다양한 시각화 기능을 제공한다.

1.라인 플롯(line plot)

2.스캐터 플롯(scatter plot)

3.컨투어 플롯(contour plot)

4.서피스 플롯(surface plot)

5.바 차트(bar chart)

6.히스토그램(histogram)

7.박스 플롯(box plot)

참고 라이브러리 : http://matplotlib.org/gallery.html

**Colormaps**

Colormap은 사용자의 데이터를 3D 공간에서 시각화하여 나타내도록 돕는데, 이 Colormap 에는 다양한 조건을 담을 수 있다. 

1.metric data 표시. 

2.데이터 집합에 대한 표시. (벗어나는 값에 대해 알 수 있다) 

3.플로팅할 매개변수에 대해 직관적으로 보여줘야 할 때.

4.보는 사람에게 표준을 제시할 수 있다. 



```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
r = 2 * np.random.rand(n)
theta = 2 * np.pi * np.random.rand(n)
area = 200 * r**2 * np.random.rand(n)
colors = theta
```
